```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
import seaborn as sns
from statsmodels.tsa import ar_model
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%matplotlib inline
```

```python
df = pd.read_csv('data3.csv')
```

```python
df.plot(figsize=(15,10))
```

```python
train = df[0:55].copy()
test = df[55:].copy()
```

# ACF and PACF

Create an ACF and PACF plot using the statsmodels library

```python
acf_plot = plot_acf(df.Sales_quantity, lags=10)
```

```python
acf_plot = plot_pacf(df.Sales_quantity, lags=10)
```

## Create and evaluate the AR model

We can use the Statsmodels tsa library (https://www.statsmodels.org/stable/tsa.html) in order to perform our next models. Please make good use of this library in order to create your AR model.

We can use the AutoReg model from this library in order to create an AR model.

```python
model = ar_model.AutoReg(endog=train.Sales_quantity, lags=3)
```

```python
res = model.fit()
```

```python
preds = res.predict(test.index[0], test.index[-1])
```

```python
test['pred'] = preds
```

```python
ax = train.plot(figsize=(10,10))
test.plot(ax=ax)
```

```python
mean_absolute_error(test['pred'], test['Sales_quantity'])
```
