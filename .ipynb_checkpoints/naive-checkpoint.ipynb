```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('temp.csv')
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
df.set_index('Date').plot(figsize=(15,10))
```

```python
train = df[0:3000].copy()
test = df[3000:].copy()
```

## Exercises:

Create the Naive, TrendNaive and Seasonal Naive algorithm and validate on the test set

### Naive

```python
Y_t = train.loc[2999].Temp
```

```python
test['pred'] = Y_t
```

```python
mean_absolute_error(test['pred'], test['Temp'])
```

```python
ax = train.set_index('Date').plot(figsize=(10,10))
test.set_index('Date').plot(ax=ax)
```

###  Trend naive

```python
t_min2 = train.loc[len(train)-2].Temp
```

```python
t_min1 = train.loc[len(train)-1].Temp
```

```python
test['pred'] = 0
```

```python
test.loc[3000, 'pred'] = Y_t.copy()
```

```python
trend = t_min1 - t_min2
```

```python
for i in test.index:
    if i == 3000:
        continue
    print(i)
    test.loc[i, 'pred'] = test.loc[i-1].pred + trend
```

```python
len(test.index)
```

```python
ax = train.set_index('Date').plot(figsize=(10,10))
test.set_index('Date').plot(ax=ax)
```

```python
mean_absolute_error(test['pred'], test['Temp'])
```

# Seasonal Naive

```python
def add_seasonal_naive(lag):
    for i in test.index:
        if i - lag < test.index[0]:
            test.loc[i, 'pred'] = train.loc[i-lag].Temp
        else:
            test.loc[i, 'pred'] = test.loc[i-lag].pred
```

```python
test['pred'] = 0 
add_seasonal_naive(365)
```

```python
ax = train.set_index('Date').plot(figsize=(10,10))
test.set_index('Date').plot(ax=ax)
```

```python
mean_absolute_error(test['pred'], test['Temp'])
```
